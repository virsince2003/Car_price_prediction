##  Data Cleaning

```python
backup=df.copy()
```

###  1) For "year" column

```python
df=df[df["year"].str.isnumeric()]
```

```python
df.shape
```

```python
df["year"]=df["year"].astype(int)
```

###  2) For "Price" column

```python
df[df["Price"]== "Ask For Price"].index
```

```python
df=df[df["Price"]!= "Ask For Price"]
```

```python
df["Price"]=df["Price"].str.replace(",","")
df["Price"]=df["Price"].astype(int)
```

###  3) For "kms_driven" column

```python
df["kms_driven"]=df["kms_driven"].str.split(" ").str.get(0).str.replace(",","")
```

```python
df=df[df["kms_driven"].str.isnumeric()]
```

```python
df["kms_driven"]=df["kms_driven"].astype(int)
```

###  4) For "fuel_type" column

```python
df=df[~df["fuel_type"].isna()]
```

```python
df["fuel_type"].unique()
```

```python
df["name"].str.split(" ").str.slice(0,3).str.join(" ")    #for name of the car
```

## Model

### Extracting Training Data

```python
x=df.drop(columns="Price")
y=df["Price"]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

## Creating an OneHotEncoder object to contain all the possible categories

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
```

## Creating a column transformer to transform categorical columns


```python
encoder=OneHotEncoder()
encoder.fit(x[['name','company','fuel_type']])
```

```python
column_trans=make_column_transformer((OneHotEncoder(categories=encoder.categories_),['name','company','fuel_type']),remainder='passthrough')
```

## Linear Regression Model

```python
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
```

## Making a pipeline

```python
pipe=make_pipeline(column_trans,regressor)
```

## Fitting the model

```python
pipe.fit(x_train,y_train)
```

```python
y_predict=pipe.predict(x_test)
```

## Checking R2 Score

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test,y_predict)
```

## Finding the model with a random state of TrainTestSplit where the model was found to give almost 0.87 as r2_scor

```python
score=[]
for i in range(2000):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.1,random_state=i)
    regressor=LinearRegression()
    pipe=make_pipeline(column_trans,regressor)
    pipe.fit(x_train,y_train)
    y_predict=pipe.predict(x_test)
    score.append(r2_score(y_test,y_predict))
```

```python
np.argmax(score)
```

```python
score[np.argmax(score)]
```

## The best model is found at a certain random state

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.1,random_state=np.argmax(score))
regressor=LinearRegression()
pipe=make_pipeline(column_trans,regressor)
pipe.fit(x_train,y_train)
y_predict=pipe.predict(x_test)
r2_score(y_test,y_predict)
```
